```python
r = np.load('results/10x10_discrete2_d2/run0/results_stan_s10000_nolookahead.pkl')
```

```python
ns, rmses, ijs, evals = map(np.asarray, zip(*r['pred-variance']))
```

```python
eval_streams = np.dstack(evals[1:-1])
n, n2, num_iters = eval_streams.shape
assert n == n2
eval_streams = eval_streams.reshape((n*n, num_iters), order='F')
```

```python
from statsmodels.graphics.boxplots import beanplot, violinplot
```

```python
useful_eval_streams = eval_streams[np.sum(np.isnan(eval_streams), axis=1) < eval_streams.shape[1] - 1,:]
useful_eval_streams.shape
```

```python
# verify that the nan pattern is some # of reals then the rest nan
for row in np.isnan(useful_eval_streams):
    nans, = np.isnan(row).nonzero()
    if nans.size:
        assert np.all(nans == np.arange(nans[0], nans[-1] + 1))
```

```python
eval_diffs = [np.diff(row[np.logical_not(np.isnan(row))]) for row in useful_eval_streams]
```

```python
# all the changes to anything
violinplot([np.hstack(eval_diffs)])
None
```

```python
# most-positive and most-negative changes per item
beanplot([[x.max() for x in eval_diffs], [x.min() for x in eval_diffs]])
None
```

```python
# get ranks of the evaluations
eval_ranks = useful_eval_streams.argsort(axis=0).argsort(axis=0).astype(float)
eval_ranks[np.isnan(useful_eval_streams)] = np.nan
```

```python
# distribution of rank changes from one step to the next
rank_changes = [np.diff(row[~np.isnan(row)]) for row in eval_ranks if row[~np.isnan(row)].size > 2]

to_plot = rank_changes[30:40]

hlines(0, 0.5, len(to_plot)+.5, color='k', linestyle='dotted')
beanplot(to_plot, plot_opts={'cutoff': True}, ax=gca())
None
```
