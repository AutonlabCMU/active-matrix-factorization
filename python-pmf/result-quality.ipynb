```python
diffs = defaultdict(list)for filename, results in zip(filenames, load_results(filenames)):    ns, random = rmse(results['random'] if 'random' in results else results['rc_random'])    for k, v in results.items():        if k != 'random' and k != 'rc_random' and not k.startswith('_'):            if 'bayes' in filename:                k = 'bayes_' + k            diffs[k].append(random - rmse(v)[1])
```

```python
def diffs_boxplot(diffs, ns, name):    boxplot(np.asarray(diffs), bootstrap=10**5, positions=ns)    plot(ns, [0] * len(ns), c='k')    title("{} vs random".format(name))    ylabel("RMSE advantage over random")    xlabel("# known")    xticks(ns, [])
```

```python
box_var = figure()diffs_boxplot(diffs['bayes_pred-variance'], ns, 'Var[R_ij]')
```
